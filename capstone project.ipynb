```python
#this is the wikipaedia url we need to work with
url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
page=urllib.request.urlopen(url)  #connecting to website
```

```python
soup=bst(page,'html.parser')
print(soup)
```

```python
#now using the regex for finding the specific parts of the page
#we need the table content which comes under class "wikitable sortable jquery-tablesorter"
regex=re.compile('^tr$')
content_list=soup.find_all(regex)
```

```python
#following string gave us all the elements starting and ending with tr tag or <tr>...</tr>
content_list
```

```python
#make a list of list by using slicing of elements 
k=[]
for i in content_list:
    i=str(i).replace('\n','')
    ls=str(i).split('</td><td>')
    if len(ls)==3:
        ls[0],ls[2]=ls[0][8:],ls[2][:len(ls[2])-10]
    k.append(ls)
```

```python
#converting to dataframe
df=pd.DataFrame(k[1:181],columns=['PostalCode','Borough','Neighborhood'])
```
